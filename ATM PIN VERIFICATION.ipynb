```python
#ðŸ” ATM PIN Verification and Withdrawal Using Nested If Statements
pin=int(input("Enter your 4-digit pin number:"))
spin= 1234
balance=5000
if pin==spin:
    print("Access Granted!!")
    withdraw=int(input("Enter amount:"))
    if withdraw>balance :
        print("Insufficent funds!!!!")
    else:
        print("Amount withdraw:", withdraw)
        print("Remaining balance:", balance-withdraw)
        print("Thanking for banking with sbi....")
else:
    print("Access decimal wrong pin")
```
